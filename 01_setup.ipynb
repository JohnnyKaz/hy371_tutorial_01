# Εισαγωγή στο Anaconda

Το <b>Anaconda</b> (ή απλά Conda) είναι ένα πολύ χρήσιμο εργαλείο για τη χρήση της Python.
Επιτρέπει τη δημιουργία εικονικών περιβαλλόντων (virtual environments) που περιέχουν sandboxed εγκαταστάσεις Python.
Με αυτόν τον τρόπο, δεν χρειάζεται να σβήνετε και να ξανακατεβάζετε την Python κάθε φορά που γίνεται κάποιο λάθος με το περιβάλλον σας.
Επιπλέον, τα περιβάλλοντα Conda επιτρέπουν την ταυτόχρονη ύπαρξη πολλαπλών εγκαταστάσεων Python, και σε κάθε μία από αυτές μπορείτε να έχετε διαφορετικές εκδόσεις Python ή διαφορετικές βιβλιοθήκες/πακέτα.

## Εγκατάσταση του Anaconda

<b>Βήμα 1</b> Για να εγκαταστήσετε το Anaconda, ακολουθήστε τον κατάλληλο οδηγό:
- Windows: https://docs.anaconda.com/anaconda/install/windows/ ή [video οδηγός](https://www.youtube.com/watch?v=eGVqBAtRqJ8)
- Linux: https://docs.anaconda.com/anaconda/install/linux/
- macOS: https://docs.anaconda.com/anaconda/install/mac-os/

Επαληθεύστε ότι η εγκατάσταση έγινε σωστά:
- Windows: Ανοίξτε το 'Anaconda Prompt' (απλά ψάξτε το στο Windows search bar). Θα πρέπει να ανοίξει ένα παράθυρο παρόμοιο με το Command Prompt, που θα γράφει όμως `(base)`.

- Linux και macOS: Ανοίξτε το terminal. Θα πρέπει να εμφανίζει τώρα `(base)` πριν το local path.

Μπορείτε να δείτε την έκδοση του `base` Python environment σας με την εντολή `python --version`.



## Jupyter και Python IDEs

Σε αυτό το φροντιστήριο (και στα επόμενα, λογικά), θα χρησιμοποιούμε Jupyter notebooks για να τρέχουμε κώδικα Python.
Τα Jupyter notebooks έχουν αρκετά πλεονεκτήματα, όπως:
- Επιτρέπουν να σπάσετε τον κώδικα σας σε ξεχωριστά κελιά (cells) και να τα τρέξετε μεμονωμένα
- Τα Cells μπορούν να περιέχουν είτε κώδικα Python είτε Markdown, το οποίο σας επιτρέπει να συνδυάσετε κώδικα με κείμενο/επεξηγήσεις/εξισώσεις με όμορφο τρόπο, πχ $e^{i\pi} + 1 = 0$

<b>Εγκατάσταση</b> Ο ευκολότερος τρόπος να εγκαταστήσετε το Jupyter είναι τρέχοντας σε κάποιο conda environment (πχ στο `base`) την εντολή

`conda install jupyter`

Όταν ολοκληρωθεί αυτό, μπορείτε να τρέξετε την εντολή

`jupyter-notebook`

για να ξεκινήσετε ένα session στο Jupyter. Αυτό θα πρέπει να ανοίξει αυτόματα ένα νέο παράθυρο στον browser σας. Αν δεν γίνει αυτό, μπορείτε να δείτε το output της εντολής στο terminal. Κάπου θα πρέπει να υπάρχει μια γραμμή που θα μοιάζει με:

`Jupyter Notebook 6.4.12 is running at: http://localhost:8889/?token=****************************`

αρκεί να αντιγράψετε αυτό το url στον browser σας.

Αν χρησιμοποιείτε ήδη κάποιο IDE όπως το PyCharm ή το VSCode για να τρέξετε κώδικα Python, μπορείτε να χρησιμοποιήσετε το Jupyter extension για να ανοίγετε τα notebooks μέσω των IDEs:
- PyCharm (μόνο στο Professional edition): https://www.jetbrains.com/help/pycharm/jupyter-notebook-support.html 
- VSCode: https://marketplace.visualstudio.com/items?itemName=ms-toolsai.jupyter
