## Εξισώσεις με latex
Μπορείς να γράψεις μαθηματικές φόρμουλες και εξισώσεις χρησιμοποιώντας latex notation (αναλυτικότερος οδηγός [εδώ](https://www.overleaf.com/learn/latex/Mathematical_expressions))

Συγκεκριμένα, βάζοντας μια εξίσωση με latex format μέσα σε `$δολλάρια$` εμφανίζεται inline, δηλαδή στην ίδια γραμμή με το κείμενο.

Για παράδειγμα, το `$\bar{x} = \frac{\sum_{i=1}^{n}x_i}{n}$` εμφανίζεται ως $\bar{x} = \frac{\sum_{i=1}^{n}x_i}{n}$

Βάζοντας μια εξίσωση μέσα σε `$2 δολλάρια$` εμφανίζεται σε ξεχωριστή γραμμή, με μεγαλύτερα γράμματα.

Για παράδειγμα, το `$$\bar{x} = \frac{\sum_{i=1}^{n}x_i}{n}$$` εμφανίζεται ως $$\bar{x} = \frac{\sum_{i=1}^{n}x_i}{n}$$

## Πίνακες
Μπορείς να προσθέσεις κάποιον πίνακα. Για να το πετύχεις αυτό, χρησιμοποιείς μια κάθετη γραμμή (`|`) για να χωρίσεις κάθε στήλη, ενώ για τον τίτλο κάθε στήλης χρειάζονται και τρεις ή περισσότερες παύλες (`---`). Μπορείτε να ευθυγραμμίσετε το κείμενο στις στήλες προς τα αριστερά, τα δεξιά ή το κέντρο, προσθέτοντας μια άνω και κάτω τελεία (`:`) στα αριστερά, στα δεξιά ή και στις δύο πλευρές των `---` μέσα στη γραμμή κεφαλίδας. Ας δούμε ένα παράδειγμα:

```
| Τίτλος 1 | Τίτλος 2   | Τίτλος 3  |
| :---     |    :----:  |      ---: |
| Αριστερή | Στοίχιση   | Δεξιά     |
| Στοίχηση | στο κέντρο | Στοίχηση  |
```
| Τίτλος 1  | Τίτλος 2   | Τίτλος 3  |
